**Understand the given variable**

**1.VIN (1-10):** The first 10 characters of the vehicle's VIN.

**2.County:** The county where the vehicle is registered.

**3.City:** The city where the vehicle is registered.

**4.State:** The state where the vehicle is registered.

**5.Postal Code:** The postal code of the vehicle's registration.

**6.Model Year:** The model year of the vehicle.

**7.Make:** The manufacturer of the vehicle.

**8.Model:** The model of the vehicle.

**9.Electric Vehicle Type:** Type of electric vehicle (e.g., Battery Electric Vehicle, Plug-in Hybrid Electric Vehicle).

**10.Clean Alternative Fuel Vehicle (CAFV) Eligibility:** Eligibility status for clean alternative fuel vehicle incentives.

**11.Electric Range:** The electric range of the vehicle in miles.

**12.Base MSRP:** Manufacturer's suggested retail price.

**13.Legislative District:** The legislative district where the vehicle is registered.

**14.DOL Vehicle ID:** Department of Licensing vehicle ID.

**15.Vehicle Location:** Geographic coordinates of the vehicle's location.

**16.Electric Utility:** The electric utility provider.

**17.2020 Census Tract:** The census tract for the vehicle's location. **Exploratory** **Data** **Analysis**


1.Geographic Distribution of Electric Vehicles:

```python
plt.figure(figsize=(10, 6))
f['Electric Vehicle Type'].value_counts().plot(kind='bar', color='skyblue')
plt.title('Electric Vehicle Distribution by Postal Code')
plt.xlabel('Postal Code')
plt.ylabel('Number of Electric Vehicles')
plt.show()
```

2.Trends in Electric Vehicle Adoption:

```python
import pandas as pd
import matplotlib.pyplot as plt
# Group by 'Model Years' and konw the electric vehicle adoption
Sales = f.groupby('Model Year')['DOL Vehicle ID'].sum().sort_values()

# Plot as a pie chart
plt.figure(figsize=(10, 6))
Sales.plot(kind='pie', autopct='%1.1f%%', colors=plt.cm.Paired.colors)
plt.title('Total Adopted Electric Vehicles')
plt.ylabel('')  # Hide the y-label for better visual appeal
plt.show()
```

3.Impact of Electric Range on EV Popularity:

```python
item_sales = f.groupby('Electric Range')['DOL Vehicle ID'].sum().sort_values()
plt.figure(figsize=(12, 8))
item_sales.plot(kind='bar', color='coral')
plt.title('Electric Range impact on EV Popularity')
plt.xlabel('DOL Vehicle ID')
plt.ylabel('Electric Range')
plt.show()
```

4.Economic Analysis of Electric Vehicle Adoption:

```python
Electric_Vehicle_Adoption_Analysis= f['Base MSRP'].value_counts()

plt.figure(figsize=(10, 6))
Electric_Vehicle_Adoption_Analysis.plot(kind='pie', autopct='%1.1f%%', colors=['skyblue', 'lightcoral'], startangle=140)
plt.title('Electric Vehicle Adoption Analysis')
plt.ylabel('')
plt.show()
```

5.Eligibility for Clean Alternative Fuel Vehicle (CAFV) Programs:

```python
#Eligibility for Clean Alternative Fuel Vehicle (CAFV) Programs
CAFV_Eligibility = f['Clean Alternative Fuel Vehicle (CAFV) Eligibility'].value_counts()

plt.figure(figsize=(10, 6))
CAFV_Eligibility.plot(kind='pie', autopct='%1.1f%%', colors=['lightgreen', 'lightcoral'], startangle=90)
plt.title('Clean Alternative Fuel Vehicle (CAFV) Eligibility')
plt.ylabel('')
plt.show()
```

6.Legislative District Analysis:

```python
#Legislative District Analysis
Legislative_District_Analysis = f['Legislative District'].value_counts()

plt.figure(figsize=(12, 6))
Legislative_District_Analysis.plot(kind='bar', color='lightblue')
plt.title('Electric Vehicle Distribution by Legislative District')
plt.xlabel('Legislative District')
plt.ylabel('Number of Electric Vehicles')
plt.xticks(rotation=90)
plt.show()
```

7.Influence of Electric Utilities:

```python
#Influence of Electric Utilities
Electric_Utilities_Influence = f['Electric Utility'].value_counts()

plt.figure(figsize=(12, 6))
Electric_Utilities_Influence.plot(kind='bar', color='lightgreen')
plt.title('Electric Vehicle Distribution by Electric Utility')
plt.xlabel('Electric Utility')
plt.ylabel('Number of Electric Vehicles')
plt.xticks(rotation=90)
plt.show()
```

8.City-Level Insights:

```python
# City-Level Insights
City_Level_Insights = f['City'].value_counts().head(10) #Considering top 10 cities for insights

plt.figure(figsize=(12, 6))
City_Level_Insights.plot(kind='bar', color='lightcoral')
plt.title('Top 10 Cities with Highest Electric Vehicle Adoption')
plt.xlabel('City')
plt.ylabel('Number of Electric Vehicles')
plt.xticks(rotation=90)
plt.show()
```

9.Electric Vehicle Type Preferences:

```python
# Electric Vehicle Type Preferences
Electric_Vehicle_Type_Preferences = f['Electric Vehicle Type'].value_counts()

plt.figure(figsize=(10, 6))
Electric_Vehicle_Type_Preferences.plot(kind='pie', autopct='%1.1f%%', colors=['lightblue', 'lightgreen'], startangle=90)
plt.title('Electric Vehicle Type Preferences')
plt.ylabel('')
plt.show()
```

10.Correlation Between MSRP and Electric Range:

```python
# Correlation Between MSRP and Electric Range:
plt.figure(figsize=(10, 6))
plt.scatter(f['Base MSRP'], f['Electric Range'], alpha=0.5)
plt.title('Correlation Between MSRP and Electric Range')
plt.xlabel('Base MSRP')
plt.ylabel('Electric Range (miles)')
plt.grid(True)
plt.show()

correlation = f['Base MSRP'].corr(f['Electric Range'])
print("Correlation coefficient:", correlation)
```
